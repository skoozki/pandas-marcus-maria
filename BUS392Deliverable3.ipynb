```python
lebron_avg = pd.read_csv("lebron_avg.csv")
lebron_avg
```

```python
jordan_avg = pd.read_csv("jordan_avgcorrect.csv")
```

```python
plt.plot(lebron_avg.Age, lebron_avg.PTS)
plt.plot(jordan_avg.Age, jordan_avg.PTS)
plt.xlabel("Age")
plt.ylabel("Points")
plt.legend(["Lebron", "Jordan"])
plt.title("Lebron Vs. Jordan Career Average in Points by Age")
plt.show()
```

```python
plt.plot(lebron_avg.Age, lebron_avg.AST)
plt.plot(jordan_avg.Age, jordan_avg.AST)
plt.xlabel("Age")
plt.ylabel("Assists")
plt.legend(["Lebron", "Jordan"])
plt.title("Lebron Vs. Jordan Career Average in Assists by Age")
plt.show()
```

```python
plt.plot(lebron_avg.Age, lebron_avg.TRB)
plt.plot(jordan_avg.Age, jordan_avg.TRB)
plt.xlabel("Age")
plt.ylabel("Total Rebounds")
plt.legend(["Lebron", "Jordan"])
plt.title("Lebron Vs. Jordan Career Average in Total Rebounds by Age")
plt.show()
```
